# Temperature Dataset

```python
months=['January','February','March','April','May']
days=np.arange(1,31)
np.random.seed(1)
val = np.random.randint(16,35,150)
readings=val.reshape(30,5)
df_temp=pd.DataFrame(data=readings,index=days, columns =months)
```

```python
df_temp['Total']= df_temp['January']+ df_temp['February']+df_temp['March']+df_temp['April']+df_temp['May']
```

```python
df_temp
```

```python
df_temp.loc['AVG']=(sum(df_temp['January']/30),sum(df_temp['February']/30),sum(df_temp['March']/30),sum(df_temp['April']/30),sum(df_temp['May']/30),sum(df_temp['Total']/30))
```

```python
df_temp
```

```python
df_temp.drop('Total', axis=1, inplace =True)
```
